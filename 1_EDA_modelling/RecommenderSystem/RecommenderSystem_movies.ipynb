```python
# RECOMMENDER SYSTEM -> input movie id, output 5 recommended movie

def content_based_recommender(movies_df, movie_id = None, top_n=5):

    # buat feature genre & desc
    movies_df['content'] = movies_df['genres'] + ' ' + movies_df['description']

    # convert text to tf-idf
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(movies_df['content'])

    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    
    idx=movies_df[movies_df['movie_id']==movie_id].index[0]


    # get similarity score
    score = list(enumerate(cosine_sim[idx]))
    score = sorted(score, key=lambda x:x[1], reverse=True)
    score = score[1:top_n+1]
    
    # get movie indices
    movie_idx = [i[0] for i in score]

    # movie recommendation
    recommendation = movies_df.iloc[movie_idx][['movie_id', 'title', 'genres', 'avg_rating']]
    recommendation['similarity_score'] = [ i[1] for i in score]

    return recommendation
```

```python
test_movie_id = 5
content_recommend = content_based_recommender(movies_df, test_movie_id)

content_recommend
```

```python
user_item_matrix = ratings_df.pivot(index='user_id', columns='movie_id',values='rating').fillna(0)
user_item_matrix
```

```python
# transpose = .T
item_sim = cosine_similarity(user_item_matrix.T, user_item_matrix.T)
```

```python
# movie yang sekiranya orang lain suka
item_sim
```

```python
# tabel 2D
# baris & kolom = nama movie
# nilai = tingkat kesamaan 2 film

movie_sim_df = pd.DataFrame(item_sim, index=user_item_matrix.columns, columns= user_item_matrix.columns)
```

```python
def item_based_recommend(movie_id=None, top_n=5):

    # cek movie ID ada di dataset/tdk
    if movie_id not in movie_sim_df.index:
        return f"Movie ID {movie_id} not found"
    
    # get similarity score
    sim_scores = movie_sim_df[movie_id].sort_values(ascending=False)[1:top_n+1]
    recommended_movies = movies_df[movies_df['movie_id'].isin(sim_scores.index)][['movie_id', 'title', 'genres', 'avg_rating']]
    recommended_movies['similarity_score'] = sim_scores.values

    return recommended_movies
```

```python
# bentuk matrix user & item 

user_item_matrix = ratings_df.pivot(index='user_id', columns='movie_id', values='rating').fillna(0)
user_item_matrix
```

```python
# 2. hitung kesamaan antar user

from sklearn.metrics.pairwise import cosine_similarity

user_sim = cosine_similarity(user_item_matrix)
user_sim_df = pd.DataFrame(user_sim, index=user_item_matrix.index, columns=user_item_matrix.index)

user_sim
```

```python
def user_based_recommendation(user_id, top_n=5):
    # cek user id ada di dataset?
    if user_id not in user_item_matrix.index:
        return f"user ID {user_id} not found in dataset"
    
    # 1. get skor kesamaan user_id
    sim_scores = user_sim_df[user_id].sort_values(ascending=False)[1:top_n+1]  # kec dirinya sendiri

    # 2. get user mirip
    sim_user = sim_scores.index[:top_n]

    # 3. ambil rating dari user yang mirip
    sim_user_ratings = ratings_df[ratings_df['user_id'].isin(sim_user)]

     # 4. hitung rata-rata rating tiap film yang belum ditonton user_id
    user_movies = ratings_df[ratings_df['user_id'] == user_id]['movie_id']
    recommendations = (
        # film blm ditonton
        sim_user_ratings[~sim_user_ratings['movie_id'].isin(user_movies)]
        .groupby('movie_id')
        .agg({'rating': 'mean'})
        .reset_index()
    )
    recommendations.columns = ['movie_id', 'rating']


    # 5. gabung dengan detail film
    recommendations = recommendations.merge(movies_df, on='movie_id')
    recommendations = recommendations.sort_values(by='rating', ascending=False).head(top_n)

    return recommendations[['movie_id', 'title', 'genres', 'rating']]
```
