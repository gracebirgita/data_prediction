```python
#NORMALISASI L1-L2
# hanya tau arah 

import numpy as np
# import pandas as pd
from sklearn.preprocessing import Normalizer

# data = pd.read_csv("titanic.csv")
# X=np.array()

#sebelum normalisasi, drop survival
target =df1['Survived']
df1 = df1.drop('Survived', axis=1)

#kolom yang akan dinormalisasi (numerik)
age_columns = df1.select_dtypes(include=np.number).columns

# Tangani nilai NaN dengan menggantinya menggunakan rata-rata kolom
df1[age_columns] = df1[age_columns].fillna(df1[age_columns].mean())
X = df1[age_columns].values



#buat L1 normalizer
normalizer = Normalizer(norm='l1')

#transform feature matrix / normalisasi data
X_normalized= normalizer.transform(X)

df1[age_columns] = X_normalized

#drop yang non numerical
categorial_features = df1.select_dtypes(include=['object', 'category']).columns
df1 = df1.drop(categorial_features, axis=1)

#masukin survived sbg target
df1['Survived']=target
```

```python
#hasil normalisasi L1-L2

print(df1[age_columns])
```

### LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
x = df1.drop('Survived', axis=1)
# x = x.drop('Name', axis=1)
y =df1['Survived']
```

```python
#split dataset

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(x_train)
```

```python
#develop LR
model =LogisticRegression()

#training
model.fit(x_train, y_train)
```

```python
#prediction
y_pred =model.predict(x_test)
```

```python
#Evaluate model : accuracy, precision, recall, ROC-AUC score


#accuracy
accuracy = accuracy_score(y_test, y_pred)

#precision, recall, f1-score, accuracy, macro avg, weighted

class_report = classification_report(y_test, y_pred)

print("Accuracy : ", accuracy)
print("classification report : ", class_report)
```

```python
#confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

*Provide a detailed analysis of the confusion matrix to understand the true positives, false positives, true negatives, and false negatives.*

Berdasarkan hasil confusion matrix

True Positive(TP) sejumlah 131

False Negative (FN) sejumlah 26

False Positive(FP) sejumlah 33

True Negative (TN) sejumlah 78 

artinya 

- model bisa memprediksi label dengan benar positive(Survived) pada dataset titanic sejumlah 131 orang dan memprediksi negative (tidak survived) sejumlah 78 orang

- model salah memprediksi label negative (tidak survived) sejumlah 26 orang dan salah memprediksi positive(survived) sejumlah 78 orang

- nilai actual data label '1' (true=survived) sejumlah 157 orang

- nilai actual data label '0' (false=not survived) sejumlah 111 orang

- Berdasarkan 164 prediksi label positive(survived) model berhasil memprediksi 131 orang dengan benar

- Berdasarkan 104 prediksi label negative(not survived) model berhasil memprediksi 78 orang dengan benar (tidak survived)

Kesimpulan : model dapat memprediksi label positive(survived -> '1') dan label negative(not survived -> '0')dengan cukup baik

```python
# Calculate ROC curve and AUC

#prediksi
y_pred = model.predict(x_test)
y_pred_prob = model.predict_proba(x_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f'ROC AUC: {roc_auc}')

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
